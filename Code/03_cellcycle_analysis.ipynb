```python
import pandas as pd
import numpy as np
import sys
import os
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype']=42
plt.rcParams['font.sans-serif']='Arial'
import warnings
warnings.filterwarnings("ignore")
```

```python
ten_data=pd.read_table('../data/10x_1993_log2_q_norm.txt',index_col=0)
ten_data
```

```python
t_cell_cluster=pd.read_table('../data/t_cell_clsuter_order.txt',index_col=0)

t_cell_cluster
```

```python
t_data_order=ten_data.T.ix[t_cell_cluster.index].T
t_data_order.dropna(axis=1,inplace=True)
t_data_order
```

```python
cell_cycle_gene=pd.read_table('../data/cellcycle_gene/GO_term_summary_20180526_091200.txt',index_col=1)
cell_cycle_gene
```

```python
T_cell_cycle=t_data_order.ix[cell_cycle_gene.index]
T_cell_cycle.dropna(inplace=True)
T_cell_cycle=T_cell_cycle.groupby('gene_id').mean()
T_cell_cycle
```

```python
marker_gene=pd.read_table('../data/marker_gene15_self_order.txt',index_col=0)
marker_gene=marker_gene[marker_gene.cluster!=13]
marker_gene=marker_gene[marker_gene.cluster!=14]
marker_gene
```

```python
T_cell_cycle_marker=T_cell_cycle.ix[marker_gene.index]
T_cell_cycle_marker.dropna(inplace=True)
T_cell_cycle_marker.drop_duplicates(inplace=True)
T_cell_cycle_marker
```

```python
mea=T_cell_cycle_marker.mean()
mea_av=mea.rolling(100,min_periods=1,center=True).mean().values
mea_av=pd.DataFrame(mea_av)
mea_all=t_data_order.mean()
mea_all_av=mea_all.rolling(100,min_periods=1,center=True).mean().values
mea_all_av=pd.DataFrame(mea_all_av)
f1=np.polyfit(np.array(mea_av.index.tolist()),mea_av.iloc[:,0].values,20)
p1 = np.poly1d(f1)
yvals1 = p1(mea_av.index.tolist())
f2=np.polyfit(np.array(mea_all_av.index.tolist()),mea_all_av.iloc[:,0].values,20)
p2 = np.poly1d(f2)
yvals2 = p2(mea_all_av.index.tolist())
plt.plot(mea_av.index.tolist(),yvals1,c='black')
plt.plot(mea_av.index.tolist(),yvals2,c='silver')
plt.show()
```

```python
T_cell_cycle_marker_aver=T_cell_cycle_marker.apply(lambda x :x.rolling(10,min_periods=1,center=True).mean(),axis=1)
T_cell_cycle_zscore=T_cell_cycle_marker.apply(lambda x : (x-T_cell_cycle_marker.mean(axis=1))/T_cell_cycle_marker.std(axis=1))
T_cell_cycle_zscore_aver=T_cell_cycle_zscore.apply(lambda x :x.rolling(10,min_periods=1,center=True).mean(),axis=1)
T_cell_cycle_zscore_aver
```

```python
sns.clustermap(T_cell_cycle_zscore_aver,col_cluster=False,row_cluster=True,metric='correlation',
               cmap='coolwarm',col_colors=t_cell_cluster.color,vmax=1,vmin=0,figsize=(10,10),
            yticklabels=False,xticklabels=False)

plt.show()
```

```python
h2m=pd.read_table('../data/mouse2human_ensemble.txt',index_col=3)
h2m=h2m.iloc[:,[1]]
h2m
```

```python
cellcluster_o=t_cell_cluster.sort_values('PBA_order')
cellcluster_o
```

```python
tdata_order=t_data_order.T.ix[cellcluster_o.index].T
g1s=pd.read_table('../data/cellcycle_gene/cellcycle4_gene_cell2015_g1s.txt',index_col=0)
gene=[i.strip() for i in g1s.index.tolist()]
g1s['gene']=gene
g1s.set_index('gene',inplace=True)
g1s=g1s.join(h2m.mouse_gene)
g1s.dropna(inplace=True)
g1s.set_index('mouse_gene',inplace=True)
g1s_data=tdata_order.ix[g1s.index]
g1s_data.dropna(inplace=True)
s=pd.read_table('../data/cellcycle_gene/cellcycle4_gene_cell2015_s.txt',index_col=0)
gene=[i.strip() for i in s.index.tolist()]
s['gene']=gene
s.set_index('gene',inplace=True)
s=s.join(h2m.mouse_gene)
s.dropna(inplace=True)
s.set_index('mouse_gene',inplace=True)
s_data=tdata_order.ix[s.index]
s_data.dropna(inplace=True)
g2m=pd.read_table('../data/cellcycle_gene/cellcycle4_gene_cell2015_g2m.txt',index_col=0)
gene=[i.strip() for i in g2m.index.tolist()]
g2m['gene']=gene
g2m.set_index('gene',inplace=True)
g2m=g2m.join(h2m.mouse_gene)
g2m.dropna(inplace=True)
g2m.set_index('mouse_gene',inplace=True)
g2m_data=tdata_order.ix[g2m.index]
g2m_data.dropna(inplace=True)
m=pd.read_table('../data/cellcycle_gene/cellcycle4_gene_cell2015_m.txt',index_col=0)
gene=[i.strip() for i in m.index.tolist()]
m['gene']=gene
m.set_index('gene',inplace=True)
m=m.join(h2m.mouse_gene)
m.dropna(inplace=True)
m.set_index('mouse_gene',inplace=True)
m_data=tdata_order.ix[m.index]
m_data.dropna(inplace=True)
mg1=pd.read_table('../data/cellcycle_gene/cellcycle4_gene_cell2015_mg1.txt',index_col=0)
gene=[i.strip() for i in mg1.index.tolist()]
mg1['gene']=gene
mg1.set_index('gene',inplace=True)
mg1=mg1.join(h2m.mouse_gene)
mg1.dropna(inplace=True)
mg1.set_index('mouse_gene',inplace=True)
mg1_data=tdata_order.ix[mg1.index]
mg1_data.dropna(inplace=True)
g1s_aver=g1s_data.mean()
s_aver=s_data.mean()
g2m_aver=g2m_data.mean()
m_aver=m_data.mean()
mg1_aver=mg1_data.mean()
cellcycle_aver=pd.concat([g1s_aver,s_aver,g2m_aver,m_aver,mg1_aver],axis=1)
cellcycle_aver=cellcycle_aver.rename(columns={0:'G1/S',1:'S',2:'G2/M',3:'M',4:"M/G1"}).T
from scipy.stats.mstats import gmean
def norm(data,method): 
    if method=="Seurat_norm":
        data=np.log(data/data.apply(lambda x:x.sum())*10000+1)
    return data
cellcycle_aver_norm=norm(cellcycle_aver.T,'Seurat_norm').T
cellcycle_aver_zscore=cellcycle_aver_norm.apply(lambda x : (x-cellcycle_aver_norm.mean(axis=1))/
                                                cellcycle_aver_norm.std(axis=1))


count_cell=[]
label=[]

for i in t_cell_cluster.stage.unique():
    b=pd.DataFrame()
    cell=cellcycle_aver_zscore.T.ix[t_cell_cluster[t_cell_cluster['stage']==i].index].T 
    for j in ['G1/S','S','G2/M','M','M/G1']:
        a=cell.T[cell.loc[j]>1]    
        frame=[b,a]
        b=pd.concat(frame)
    b.drop_duplicates(inplace=True)
    stage=[]

    for k in b.index:
        stage.append(b.ix[k].idxmax())
        label.append(k+'\t'+b.ix[k].idxmax())
    from collections import Counter
    c=Counter(stage)
    for key in c:
        if len(c)>-1:
            count_cell.append(str(i)+'\t'+key+'\t'+str(float(c[key])/cell.shape[1]))
d={}
import collections
d = collections.defaultdict(dict)
for line in count_cell:
    item=line.split('\t')
    d[item[0]][item[1]]=float(item[2])
pdata_z=pd.DataFrame(d)
pdata_z=pdata_z.T
pdata_z=pdata_z.T
pdata_z['4DPre2']=0
pdata_z=pdata_z.T
pdata_z['NO']=1-pdata_z.sum(axis=1)
pdata_z.sort_index(inplace=True)
pdata_z.fillna(0,inplace=True)
pdata_z
plt.figure(figsize=(10,4))
x=pdata_z.index.tolist()
y1=np.array(pdata_z['G1/S'].values.tolist())
y2=np.array(pdata_z['S'].values.tolist())
y3=np.array(pdata_z['G2/M'].values.tolist())
y4=np.array(pdata_z['M'].values.tolist())
y5=np.array(pdata_z['M/G1'].values.tolist())
y6=np.array(pdata_z['NO'].values.tolist())
plt.bar(x, y1, color='forestgreen', label='G1/S')
plt.bar(x, y2, bottom=y1, color='dodgerblue', label='S')
plt.bar(x, y3, bottom=y1+y2, color='rebeccapurple', label='G2/M')
plt.bar(x, y4, bottom=y1+y2+y3, color='r', label='M')
plt.bar(x, y5, bottom=y1+y2+y3+y4, color='gold', label='M/G1')
plt.bar(x, y6, bottom=y1+y2+y3+y4+y5, color='lightgray', label='nocellcycle')
plt.figlegend()

plt.show()
```
